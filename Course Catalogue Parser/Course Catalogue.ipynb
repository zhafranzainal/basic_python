```python
import camelot
tables = camelot.read_pdf('COURSE_CATALOG_IJA.pdf', pages='all')
tables[0].df = tables[0].df.replace('\\n',' ', regex=True)
tables[0].df
```

```python
tables[0].df.iloc[1, :6] = tables[0].df.iloc[1, 0].split()[:6]
tables[0].df
```

```python
modes = tables[0].df.iloc[3, 4].split()[:8]

for i in range(2, 6):
    tables[0].df.iloc[i, 4] = modes[2 * (i - 2)] + modes[2 * (i - 2) + 1]

tables[0].df
```
